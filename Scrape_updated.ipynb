```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob

ds=pd.read_excel("Sample websites.xlsx")
s=[]
f=[]
g=[]
t=[]
s=ds['homepage'].values
```

```python
for i in s:
    url = i
    print(url)
    try:
        page = requests.get(url)        
        html_code = page.content        
    except Exception as e:
        print(e)
    try:
        soup = BeautifulSoup(html_code,'html.parser')
        try:
            meta=soup.find('meta',attrs={"name":"description"})   
            
    #         print(meta.attrs['content'])
            blob = TextBlob(meta.attrs['content'])
            Englishstr = blob.translate(to="EN")
            print(Englishstr)
            f.append(Englishstr)
            title = soup.select('head title')
            title = [t.text for t in title]
            title = ' '.join(title)
#             print(title)
            blob1 = TextBlob(title)
            Englishstr1 = blob1.translate(to="EN")
            print(Englishstr1)
            g.append(Englishstr1)
        except:
            title = soup.select('head title')
            title = [t.text for t in title]
            title = ' '.join(title)
#             print(title)
            blob1 = TextBlob(title)
            Englishstr1 = blob1.translate(to="EN")
            print(Englishstr1)
            g.append(Englishstr1)
            f.append("")

    except Exception as e:
        print(e)
        f.append("")
        g.append("")
#         f.append("")

#     print(title)


'''for i in range(f):
    if "service" in f[i] or "service" in g[i]:
        t[i]="authorised car service"
    elif "tyre" in f[i] or "tyre" in g[i]:
        t[i]="tyre service"
    elif ("glass" or "stone") in f[i] or ("glass" or "stone")  in g[i]:
        t[i]="glass service"
    elif ("body" or "paint")  in f[i] or  ("body" or "paint")  in g[i]:
        t[i]="body and paint service"
    else:
        t[i]="independent car service"
df = pd.DataFrame(list(zip(s, t)), 
       columns =['url', 'target_groups'])'''
```

```python

t={}
for i in range(len(s)):
    if ("glass" or "stone") in str(f[i]) or ("glass" or "stone")  in str(g[i]):
        t[s[i]]="glass service"
    elif "tyre" in str(f[i])or "tyre" in str(g[i]):
        t[s[i]]="tyre service"
    elif "service" in str(f[i]) or "service" in str(g[i]):
        t[s[i]]="authorised car service"
    elif ("body" or "paint")  in str(f[i]) or  ("body" or "paint")  in str(g[i]):
        t[s[i]]="body and paint service"
    else:
        t[s[i]]="independent car service"
```

```python
pd.DataFrame(t.items(),columns =['url', 'target_groups'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
p=[]
for i in g:
    p.append(str(i))
l=ds['target_groups'].values
model.fit(p[:130],l[:130])
labels=model.predict(p[160:])
dz = pd.DataFrame(list(zip(s[160:], labels)), columns =['url', 'target_groups'])
dz.to_excel('new.xlsv')
```
